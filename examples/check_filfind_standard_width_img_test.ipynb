## how setting the standard width in filfinder (new) change the output results

```python
from astropy.io import fits
from fil_finder import FilFinder2D
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from astropy import units as u
```

```python
import numpy as np
```

```python
ex_slice_file = '/Volumes/LarryExternal1/Research_2017/GALFA_slices_backup/umask_gaussian_30/GALFA_HI_W_S0955_V-050.4kms_umask.fits'
data, hdr = fits.getdata(ex_slice_file, header=True)
```

```python
hdr['BUNIT'] = 'k'
```

```python
data.shape
```

```python
plt.imshow(data, origin='lower', cmap='gray')
```

```python
data_cut = data[500:1500, 9000:10000]
```

```python
plt.imshow(data_cut, origin='lower', cmap='binary')
```

```python
fil = FilFinder2D(data_cut, header=hdr,
                  distance=100.*u.pc,beamwidth=10.*u.arcmin)
```

```python
fil.preprocess_image(flatten_percent=95)
```

```python
plt.imshow(fil.flat_img.value, origin='lower', cmap='binary')
```

```python
s_w = 0.1 * u.pc
fil.create_mask(smooth_size=s_w/2,
                size_thresh=8*s_w**2*4, #6*FWHM * FWHM * pi  # 6*FWHM*2 * 2*FWHM
                adapt_thresh=s_w*2,
                verbose=True, border_masking=False)
```

```python
s_w = 0.1 * u.pc
fil.create_mask(smooth_size=s_w/2,
                size_thresh=2*s_w**2*np.pi, #6*FWHM * FWHM * pi  # 6*FWHM*2 * 2*FWHM
                adapt_thresh=s_w*2,
                verbose=True, border_masking=False)
```

```python
s_w = 0.24 * u.pc
fil.create_mask(smooth_size=s_w/2,
                size_thresh=6*s_w**2*np.pi,
                adapt_thresh=s_w*2,
                verbose=True, border_masking=False)
```

```python
s_w = 0.3 * u.pc
fil.create_mask(smooth_size=s_w/2,
                size_thresh=5*s_w**2*np.pi,
                adapt_thresh=s_w*2,
                verbose=True, border_masking=False)
```

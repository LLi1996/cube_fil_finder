```python
import pickle
import os
import glob
```

```python
import logging
logging.getLogger().setLevel('INFO')
```

```python
base_dir = '/Volumes/LarryExternal1/Research_2017/pickled_dicts_backup/new_filfind_1/'
```

```python
old_node_p_files = glob.glob(base_dir + '*')
```

```python
ex_node_p_file = old_node_p_files[0]
```

```python
nodes = pickle.load(open(ex_node_p_file, 'rb'))
```

```python
from cube_fil_finder.structs import mask_obj_node
```

```python
def convert_nodes_to_new_nodes(nodes):
    new_nodes = dict()
    for k in nodes:
        node = nodes[k]
        new_node = mask_obj_node.MaskObjNode(node.mask, node.corners_original, node.v_slice_index[0])
        new_nodes[k] = new_node
    return new_nodes
```

```python
for old_node_p_file in old_node_p_files:
    logging.info('on {0}'.format(old_node_p_file))
    new_node_p_file = old_node_p_file.replace('new_filfind_1', 'new_filfind_new_nodes')
    logging.info('loading {0} ...'.format(old_node_p_file))
    old_nodes = pickle.load(open(old_node_p_file, 'rb'))
    logging.info('converting to new node structure ...')
    new_nodes = convert_nodes_to_new_nodes(old_nodes)
    logging.info('saving to {0} ...'.format(new_node_p_file))
    pickle.dump(new_nodes, open(new_node_p_file, 'wb'))
```

```python
import sys
import glob
import pickle
```

```python
def fix_corners(node, img_dim):
    m, n = img_dim
    m_mask, n_mask = node.mask.shape
    corners = node.corners_original
    assert(corners[0][0] == node.corner_BL[1])
    assert(corners[0][1] == node.corner_BL[0])
    assert(corners[1][0] == node.corner_TR[1])
    assert(corners[1][1] == node.corner_TR[0])
    assert(corners[0][0] == node.corners[0][1])
    assert(corners[0][1] == node.corners[0][0])
    assert(corners[1][0] == node.corners[1][1])
    assert(corners[1][1] == node.corners[1][0])
    m_mismatch = corners[1][0] - corners[0][0] - m_mask
    n_mismatch = corners[1][1] - corners[0][1] - n_mask
    if m_mismatch != 0 or n_mismatch != 0:
        print('\t\tfound issue')
        print('\t\t\t mask shape: {0}'.format(node.mask.shape))
        print('\t\t\t corners: {0}'.format(corners))
        if corners[0][0] == -1 or corners[0][1] == -1:
            # fix bad bug, width comp test only
            if m_mismatch > 0 and corners[0][0] == -1:
                corners[0][0] = 0
            if n_mismatch > 0 and corners[0][1] == -1:
                corners[0][1] = 0
        else:
            if m_mismatch < 0:
                if corners[0][0] == 0:
                    corners[0][0] = -1
                if corners[1][0] == m - 1:
                    corners[1][0] = m
            if n_mismatch < 0:
                if corners[0][1] == 0:
                    corners[0][1] = -1
                if corners[1][1] == n - 1:
                    corners[1][1] = n
        print('\t\t\t new corners: {0}'.format(corners))

    node.corner_BL = [corners[0][1], corners[0][0]]
    node.corner_TR = [corners[1][1], corners[1][0]]
    node.corners = [node.corner_BL, node.corner_TR]
    node.corners_original = corners
        

def batch_fix_corners(parent_dir, img_dim):
    """
    Args:
        parent_dir {str} -- directory
        img_dim {tuple} -- of shape (m by n)
    """
    node_pickle_files = sorted(glob.glob(parent_dir + '*.p'))
    print('found {0} pickle files'.format(len(node_pickle_files)))
    
    for f in node_pickle_files:
        print('\tloading {0}'.format(f))
        nodes = pickle.load(open(f, 'rb'))
        for k in nodes:
            node = nodes[k]
            fix_corners(node, img_dim)
        print('\tsaving {0}'.format(f))
        pickle.dump(nodes, open(f, 'wb'))
```

```python
parent_dir = '../pickled_dicts/filfind_width_comp_high_1/'
```

```python
batch_fix_corners(parent_dir, (1000, 1000))
```

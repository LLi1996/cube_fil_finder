```python
import pickle
import cube_processes
import glob
from astropy.io import fits

import sys
fil_finder_dir = '/Users/larryli/Documents/CC/16-17/research/GALFA_filfind/fil_finder'
sys.path.append(fil_finder_dir)
import filfind_class as filfind
```

```python
%matplotlib inline
```

```python
slice_dir_15 = '../data/umask_from_susan/_15/'
slice_dir_30 = '../data/umask_from_susan/_30/'

slice_common_name = 'GALFA_HI_W_S1011_V-009.2kms_umask.fits'
```

```python
data_15, hdr_15 = fits.getdata(slice_dir_15 + slice_common_name, header=True)
data_30, hdr_30 = fits.getdata(slice_dir_30 + slice_common_name, header=True)
```

```python
x_range = [7000, 8000]
y_range = [1000, 2000]

data_15_test = data_15[y_range[0]:y_range[1], x_range[0]:x_range[1]]
data_30_test = data_30[y_range[0]:y_range[1], x_range[0]:x_range[1]]
```

```python
fils_15 = filfind.fil_finder_2D(data_15_test, header=hdr_30, beamwidth=10.0, glob_thresh=20,
                                 distance=100, flatten_thresh=95, standard_width=1,
                                 size_thresh=600)
# note size_thresh, adapt_thresh, smooth_size, fill_hole_size can all be set by args
mask_objs_15 = fils_15.create_mask(verbose=True, regrid=False, border_masking=True,
                                 save_png=True, run_name=str(1011), output_mask_objs=True,
                                 test_mode=False)
```

```python
fils_30 = filfind.fil_finder_2D(data_30_test, header=hdr_30, beamwidth=10.0, glob_thresh=20,
                                 distance=100, flatten_thresh=95, standard_width=1,
                                 size_thresh=600)
# note size_thresh, adapt_thresh, smooth_size, fill_hole_size can all be set by args
mask_objs_30 = fils_30.create_mask(verbose=True, regrid=False, border_masking=True,
                                 save_png=True, run_name=str(1011), output_mask_objs=True,
                                 test_mode=False)
```

```python
fils_15_newpara = filfind.fil_finder_2D(data_15_test, header=hdr_30, beamwidth=10.0, glob_thresh=20,
                                 distance=100, flatten_thresh=95, standard_width=0.2,
                                 size_thresh=600)
# note size_thresh, adapt_thresh, smooth_size, fill_hole_size can all be set by args
mask_objs_15_newpara = fils_15_newpara.create_mask(verbose=True, regrid=False, border_masking=True,
                                 save_png=True, run_name=str(1011), output_mask_objs=True,
                                 test_mode=False)
```

```python
fils_30_newpara = filfind.fil_finder_2D(data_30_test, header=hdr_30, beamwidth=10.0, glob_thresh=20,
                                 distance=100, flatten_thresh=95, standard_width=.2,
                                 size_thresh=600)
# note size_thresh, adapt_thresh, smooth_size, fill_hole_size can all be set by args
mask_objs_30_newpara = fils_30_newpara.create_mask(verbose=True, regrid=False, border_masking=True,
                                 save_png=True, run_name=str(1011), output_mask_objs=True,
                                 test_mode=False)
```

```python
def vis_tree_shadow(tree, save_fig=False, save_dir=None, save_name=None):
    root = tree.root_node
    starting_v = tree.root_v_slice
    length = tree.length
    assert(tree.has_ended)
    
    if save_fig and save_name is None:
        save_name = ''.format()
    
    plot_corners = [root.corner_BL[0], root.corner_TR[0], root.corner_BL[1], root.corner_TR[1]]

    plot = plt.imshow([[],[]], extent=plot_corners)
    plot = plt.contour(root.mask, colors='r', extent=plot_corners)
    
    plt.savefig(save_dir + save_name)
```

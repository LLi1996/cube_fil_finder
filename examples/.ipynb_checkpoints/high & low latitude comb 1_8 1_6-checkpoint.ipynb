### Combining the low and high latitude 1_8 cut trees

```python
import pickle
from cube_fil_finder.structs import util as struct_util
from cube_fil_finder.vis import tree_vis
%matplotlib inline
reload(tree_vis)
```

```python
pickle_path_high_b_1_8 = '../pickled_dicts/third_batch/all_trees_1_8.p'
pickle_path_low_b_1_8 = '../pickled_dicts/second_batch/chosen_all_trees_1_8.p'
trees_high_b_1_8 = pickle.load(open(pickle_path_high_b_1_8, 'rb'))
trees_low_b_1_8 = pickle.load(open(pickle_path_low_b_1_8, 'rb'))
```

```python
print('{0} trees in high latitude 1:8 cut, {1} trees in low latitude 1:8 cut'
      .format(len(trees_high_b_1_8), len(trees_low_b_1_8)))
```

```python
trees_all_sky_1_8 = trees_high_b_1_8
for k in trees_low_b_1_8:
    tree = trees_low_b_1_8[k]
    struct_util.add_tree_to_dict(tree, trees_all_sky_1_8)
```

```python
print('{0} trees in the combined all sky 1:8 set'.format(len(trees_all_sky_1_8)))
```

```python
pickle_path_all_sky_1_8 = '../pickled_dicts/all_sky_1_8_cut/all_sky_1_8.p'
pickle.dump(trees_all_sky_1_8, open(pickle_path_all_sky_1_8, 'wb'))
```

```python
fig1 = tree_vis.vis_trees_sky_dist(trees_all_sky_1_8, 'All Sky 1:8', return_fig=True)
fig1.show()
fig1.savefig('../vis/all_sky_maps/all_sky_1_8_dist.pdf')
```

### Combining the low and high latitude 1_6 cut trees

```python
pickle_path_high_b_1_6 = '../pickled_dicts/third_batch/all_trees_1_6.p'
pickle_path_low_b_1_6 = '../pickled_dicts/second_batch/chosen_all_trees_1_6.p'
trees_high_b_1_6 = pickle.load(open(pickle_path_high_b_1_6, 'rb'))
trees_low_b_1_6 = pickle.load(open(pickle_path_low_b_1_6, 'rb'))

print('{0} trees in high latitude cut, {1} trees in low latitude cut'
      .format(len(trees_high_b_1_6), len(trees_low_b_1_6)))

trees_all_sky_1_6 = trees_high_b_1_6
for k in trees_low_b_1_6:
    tree = trees_low_b_1_6[k]
    struct_util.add_tree_to_dict(tree, trees_all_sky_1_6)
    
print('{0} trees in the combined all sky 1:6 set'.format(len(trees_all_sky_1_6)))

pickle_path_all_sky_1_6 = '../pickled_dicts/all_sky_1_6_cut/all_sky_1_6.p'
pickle.dump(trees_all_sky_1_6, open(pickle_path_all_sky_1_6, 'wb'))
```

```python
fig2 = tree_vis.vis_trees_sky_dist(trees_all_sky_1_6, 'All Sky 1:6', return_fig=True)
fig2.show()
```

```python
fig2.savefig('../vis/all_sky_maps/all_sky_1_6_dist.pdf')
```

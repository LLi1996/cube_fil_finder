# try k means

```python
import numpy as np
from scipy.cluster.vq import kmeans2
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
data_file_path = '../data_out/all_prop.txt'
```

```python
# all the data fields
keys = ['ra',
        'dec',
        'l',
        'b',
        'size',
        'roundness',
        'starting_velocity',
        'velocity_span',
        'average_velocity',
        'moment_0_mean',
        'moment_0_median',
        'moment_0_peak',
        'moment_1_mean',
        'moment_1_median',
        'moment_2_mean',
        'moment_2_median',
        'moment_2_peak',
        'column_density_mean',
        'column_density_median',
        'column_density_peak',
        'width_fit',
        'width_err']
```

```python
data = np.loadtxt(data_file_path, delimiter=',')
```

```python
data.shape # 533 trees (excluding edge cases and ones that are too long in velocity), 22 fields outlined in keys
```

```python
ras = data[:, 0]
decs = data[:, 1]
ls = data[:, 2]
bs = data[:, 3]
sizes = data[:, 4]
roundnesses = data[:, 5]
starting_velocities = data[:, 6]
velocity_spans = data[:, 7]
average_velocities = data[:, 8]
moment_0_means = data[:, 9]
moment_0_medians = data[:, 10]
moment_0_peaks = data[:, 11]
moment_1_means = data[:, 12]
moment_1_medians = data[:, 13]
moment_2_means = data[:, 14]
moment_2_medians = data[:, 15]
moment_2_peaks = data[:, 16]
column_density_means = data[:, 17]
column_density_medians = data[:, 18]
column_density_peaks = data[:, 19]
width_fits = data[:, 20]
width_errs = data[:, 21]
```

```python
idx_sans_nans = np.where(np.isfinite(width_fits))
```

```python
cluster_data = np.array([np.abs(bs[idx_sans_nans]),
                         np.abs(moment_1_medians[idx_sans_nans]),
                         width_fits[idx_sans_nans]]).T
```

```python
centroid, label = kmeans2(cluster_data, 2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.scatter(cluster_data[:, 1][label==1], cluster_data[:, 0][label==1], color='red')
ax.scatter(cluster_data[:, 1][label==0], cluster_data[:, 0][label==0], color='blue')

ax.set_ylim(0, 90)
ax.set_xlim(0, 55)
ax.set_ylabel(r'$|b|\ [^\circ]$', fontsize=18)
ax.set_xlabel(r'$\mathrm{median\ } v\  [km\ s^{-1}]$', fontsize=18)
fig.tight_layout()
fig.show()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.scatter(cluster_data[:, 2][label==1], cluster_data[:, 0][label==1], color='red')
ax.scatter(cluster_data[:, 2][label==0], cluster_data[:, 0][label==0], color='blue')

ax.set_ylim(0, 90)
ax.set_xlim(0.1)
ax.set_ylabel(r'$|b|\ [^\circ]$', fontsize=18)
ax.set_xlabel(r'$\mathrm{median\ } v\  [km\ s^{-1}]$', fontsize=18)
fig.tight_layout()
fig.show()
```

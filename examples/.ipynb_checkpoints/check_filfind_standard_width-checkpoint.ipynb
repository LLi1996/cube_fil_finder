```python
def filter_trees(trees, aspect='1_6', size=0, v_length=1):
    filtered_trees = {}
    for k in trees:
        this_tree = trees[k]
        tree_roundness = moments.get_tree_mask_orientation_info(trees[k])[4]
        tree_v_length = this_tree.length
        tree_masked_size = this_tree.getTreeMaskedArea2D()
        if tree_roundness < moments.ROUNDNESS_AR_CONVERSION[aspect] and tree_v_length > v_length and tree_masked_size > size:
            filtered_trees[k] = this_tree
    return filtered_trees
```

```python
filtered_trees_h_1 = filter_trees(all_trees_h_1)
filtered_trees_h_25 = filter_trees(all_trees_h_25)
filtered_trees_h_5 = filter_trees(all_trees_h_5)
filtered_trees_l_1 = filter_trees(all_trees_l_1)
filtered_trees_l_25 = filter_trees(all_trees_l_25)
filtered_trees_l_5 = filter_trees(all_trees_l_5)
```

```python
def find_matches(thin, wider, widest, overlap_thresh=.85):
    wdst_counter = 0
    wdst_wder_counter = 0
    wder_counter = 0
    if len(widest) > 0:
        for kwdst in widest:
            tree_wdst = widest[kwdst]
            tree_wdst_a = tree_wdst.getTreeMaskedArea2D()
            t_wdst_found = False
            t_wdst_wder_found = False
            for kwder in wider:
                tree_wder = wider[kwder]
                tree_wder_a = tree_wder.getTreeMaskedArea2D()
                t_wder_found = False
                for kt in thin:
                    tree_t = thin[kt]
                    tree_t_a = tree_t.getTreeMaskedArea2D()
                    if not t_wder_found and tree_t.root_node.checkMaskOverlap(tree_wder.root_node, overlap_thresh):
                        wder_counter += 1
                        t_wder_found = True
                    if not t_wdst_found and tree_t.root_node.checkMaskOverlap(tree_wdst.root_node, overlap_thresh):
                        wdst_counter += 1
                        t_wdst_found = True
                    if not t_wdst_wder_found and tree_wder.root_node.checkMaskOverlap(tree_wdst.root_node, overlap_thresh):
                        #if .1 * tree_wder_a > 1.* abs(tree_wdst_a - tree_wder_a) /tree_wder_a: 
                        wdst_wder_counter += 1
                        t_wdst_wder_found = True               
    else:
        for kwder in wider:
            tree_wder = wider[kwder]
            tree_wder_a = tree_wder.getTreeMaskedArea2D()
            t_wder_found = False
            for kt in thin:
                tree_t = thin[kt]
                tree_t_a = tree_t.getTreeMaskedArea2D()
                if not t_wder_found and tree_t.root_node.checkMaskOverlap(tree_wder.root_node, overlap_thresh):
                    wder_counter += 1
                    t_wder_found = True          
    print('{0} out of {1} of widest trees found in wider trees'.format(wdst_wder_counter, len(widest)))
    print('{0} out of {1} of widest trees found in thin trees'.format(wdst_counter, len(widest)))
    print('{0} out of {1} of wider trees found in thin trees'.format(wder_counter, len(wider)))    
```

```python
def vis_width_comp(thin, wider, widest):
    fig, axs = plt.subplots(nrows=1, ncols=3, figsize=(12,5))
    ax1, ax2, ax3 = axs
    labels = ['filfind FHWM = 0.1pc', 'filfind FHWM = 0.25pc', 'filfind FHWM = 0.5pc']

    for key in thin:
        this_node = thin[key]
        this_mask = this_node.root_node.mask
        this_plot_corners = [this_node.root_node.corner_BL[0], this_node.root_node.corner_TR[0],
                             this_node.root_node.corner_BL[1], this_node.root_node.corner_TR[1]]
        ax1.contour(this_mask, colors='r', extent=this_plot_corners, linewidths=.3)

    for key in wider:
        this_node = wider[key]
        this_mask = this_node.root_node.mask
        this_plot_corners = [this_node.root_node.corner_BL[0], this_node.root_node.corner_TR[0],
                             this_node.root_node.corner_BL[1], this_node.root_node.corner_TR[1]]
        ax2.contour(this_mask, colors='r', extent=this_plot_corners, linewidths=.3)

    for key in widest:
        this_node = widest[key]
        this_mask = this_node.root_node.mask
        this_plot_corners = [this_node.root_node.corner_BL[0], this_node.root_node.corner_TR[0],
                             this_node.root_node.corner_BL[1], this_node.root_node.corner_TR[1]]
        ax3.contour(this_mask, colors='r', extent=this_plot_corners, linewidths=.3)

    for i, ax in enumerate(axs):
        ax.set_aspect('equal')
        ax.set_title(labels[i])
        ax.imshow([[], []], extent=[0, 1000, 0, 1000], origin='lower')
    fig.tight_layout()
    fig.show()
    return fig
```

```python
import pickle
import os
import glob
from matplotlib import pyplot as plt

from cube_fil_finder.util import tree_dict_util
from cube_fil_finder.structs import mask_obj_node

%matplotlib inline
```

```python
debug_tree_dict_dir = '../../debug/avery_test_nodes'
debug_tree_dict_dir = os.path.abspath(debug_tree_dict_dir)
```

```python
debug_tree_dict_paths = glob.glob(os.path.join(debug_tree_dict_dir, '*')) 
```

```python
nodes_dicts = dict()
for path in debug_tree_dict_paths:
    _, f_name = os.path.split(path)
    f_name, _ = os.path.splitext(f_name)
    nodes_slice_name = str(int(f_name[9:]))
    with open(path, 'rb') as f:
        nodes_dict = pickle.load(f)
        nodes_dicts[nodes_slice_name] = nodes_dict
```
